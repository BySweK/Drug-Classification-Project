```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('drug200.csv')
```

```python
data.head()
```

```python
#BP: Blood Pressure Levels
#Na_to_K : Na to Potassium Ration
```

```python
data.info()
```

```python
data.describe()
```

```python
# Korelasyona sokmak için Sex,BP,Chloesterol sütunlarını encode etmemiz gerekiyor.
```

```python
#Sex : Cinsiyet Kolonu Label Encode
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
data["Sex"] = encoder.fit_transform(data["Sex"])
```

```python
data["Sex"].unique()
```

```python
#BP : Kan Basıncı Kolonu Ordinal Encode
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data["BP"].unique()
```

```python
siralama = [['LOW','NORMAL','HIGH']]
```

```python
encoder2 = OrdinalEncoder(categories=siralama)
```

```python
data['BP'] = encoder2.fit_transform(data[['BP']])
```

```python
data["BP"].unique()
```

```python
#Cholesterol : Kolesterol Kolonu Ordinal Encode
```

```python
data["Cholesterol"].unique()
```

```python
siralama2 = [['NORMAL','HIGH']]
```

```python
encoder3 = OrdinalEncoder(categories=siralama2)
```

```python
data['Cholesterol'] = encoder3.fit_transform(data[['Cholesterol']])
```

```python
data.describe()
```

```python
data["Drug"].unique()
```

```python
# Y degeri için Label Encode
```

```python
encoder4 = LabelEncoder()
```

```python
data["Drug"] = encoder4.fit_transform(data["Drug"])
```

```python
data["Drug"].unique()
```

```python
# Verilerin uyumu ve doğru gradient işlemi için scale işlemi
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop("Drug",axis=1)
y = data["Drug"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression Import
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(X_train_scaled,y_train)
y_pred = reg.predict(X_test_scaled)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("Result:",accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
# LogisticRegression : 0.94
```

```python
# SVM İle Aynı Datayı Test Etme İşlemi
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC() #kernel: rbf olarak default
```

```python
svc.fit(X_train_scaled,y_train)
y_pred = svc.predict(X_test_scaled)
print("Result:",accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
# SVC ile biraz daha uygun sonuç elde ettim. Biraz da grid işlemleri ile arttırmaya çalıştım.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C' : [0.1,1,10,100,1000],
    "gamma" : ['scale','auto'],
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid']
}
```

```python
grid = GridSearchCV(SVC(),param_grid=params,cv=5,n_jobs=-1)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test_scaled)
print("Result:",accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
# Logistic Regresyon ile 0.94, SVC ile 0.96, hiperparametre(grid) ile birlikte 0.98 sonuç elde ettim.
```
